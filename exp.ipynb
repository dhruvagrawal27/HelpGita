```python
def load_pdf(data):
    loader = DirectoryLoader(data,
                    glob="*.pdf",
                    loader_cls=PyPDFLoader)
    
    documents = loader.load()

    return documents
```

```python
import os
from pinecone import Pinecone
from langchain.embeddings import OpenAIEmbeddings
from dotenv import load_dotenv
from typing import List

load_dotenv()
PINECONE_API_KEY = os.getenv('PINECONE_API_KEY')
```

```python
pc = Pinecone(api_key=PINECONE_API_KEY)
index = pc.Index("helpgita")  # replace with your index name
```

```python
from langchain.schema import Document
def store_embeddings(documents: List[Document], namespace: str = "default") -> None:
    """
    Create embeddings from documents and store them in Pinecone.
    
    Args:
        documents: List of Document objects
        namespace: Pinecone namespace to store vectors
    """
    # Extract text content from Document objects
    texts = [doc.page_content for doc in documents]
    
    # Generate embeddings for all texts
    text_embeddings = embeddings.embed_documents(texts)
    
    # Prepare vectors for upsert
    vectors = []
    for i, (doc, embedding) in enumerate(zip(documents, text_embeddings)):
        vectors.append({
            "id": f"doc_{i}",
            "values": embedding,
            "metadata": {
                "text": doc.page_content,
                # Include any additional metadata from the Document
                **doc.metadata
            }
        })
    
    # Upsert vectors in batches
    batch_size = 100
    for i in range(0, len(vectors), batch_size):
        batch = vectors[i:i + batch_size]
        index.upsert(vectors=batch, namespace=namespace)
    
    print(f"Stored {len(texts)} documents in Pinecone namespace: {namespace}")
```

```python
def similarity_search(query: str, namespace: str = "default", k: int = 3) -> List[Document]:
    """
    Perform similarity search for a query string.
    
    Returns:
        List of Document objects
    """
    # Generate embedding for query
    query_embedding = embeddings.embed_query(query)
    
    # Query Pinecone
    results = index.query(
        namespace=namespace,
        vector=query_embedding,
        top_k=k,
        include_metadata=True
    )
    
    # Convert results back to Document objects
    documents = []
    for match in results.matches:
        # Create Document with original text and metadata
        doc = Document(
            page_content=match.metadata["text"],
            metadata={k: v for k, v in match.metadata.items() if k != "text"}
        )
        documents.append(doc)
    
    return documents
```

```python

store_embeddings(text_chunks, namespace="HelpGita")
```

```python
query = "What should I do if I am an introvert and have fear of speaking to strangers?"

# Search for similar texts
results = similarity_search(query, namespace="HelpGita", k=3)

# Print results
print("Query:", query)
print("\nSimilar texts:")
for i, text in enumerate(results, 1):
    print(f"\n{i}. {text}")
```
